```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

```python
# Line plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x,y,z)
```

```python
# Rotating camera angles: elevation & azimuth
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x,y,z)
ax.view_init(azim=180)
```

```python
# Scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x,y = np.meshgrid(np.arange(-10,10,0.5), np.arange(-10,10,0.5))
r = np.linalg.norm([x,y], axis=0)
goldstone = -160*np.power(r,2)+np.power(r,4)
ax.scatter(x,y,goldstone, s=1)
```

```python
# Wireframe plot: r/c, stride & count
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x,y = np.meshgrid(np.arange(-10,10,0.5), np.arange(-10,10,0.5))
r = np.linalg.norm([x,y], axis=0)
goldstone = -160*np.power(r,2)+np.power(r,4)
ax.plot_wireframe(x,y,goldstone, rstride=10)
```

```python
# Surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x,y = np.meshgrid(np.arange(-10,10,0.5), np.arange(-10,10,0.5))
r = np.linalg.norm([x,y], axis=0)
goldstone = -160*np.power(r,2)+np.power(r,4)
ax.plot_surface(x,y,goldstone, rstride=1, cstride=1)
```

```python
# Surface plot: appearance w/ color, cmap
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x,y = np.meshgrid(np.arange(-10,10,0.5), np.arange(-10,10,0.5))
r = np.linalg.norm([x,y], axis=0)
goldstone = -160*np.power(r,2)+np.power(r,4)
ax.plot_surface(x,y,goldstone, rstride=1, cstride=1, cmap='viridis')
```

```python
# Tri-surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x,y = np.meshgrid(np.arange(-10,10,0.5), np.arange(-10,10,0.5))
r = np.linalg.norm([x,y], axis=0)
goldstone = -100*np.power(r,2)+np.power(r,4)
x = x[r<10]
y = y[r<10]
goldstone = goldstone[r<10]
ax.plot_trisurf(x,y,goldstone, color='none')
```

```python
# Contour plot (filled/unfilled)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x,y = np.meshgrid(np.arange(-10,10,0.5), np.arange(-10,10,0.5))
r = np.linalg.norm([x,y], axis=0)
goldstone = -200*np.power(r,2)+np.power(r,4)
ax.contourf(x,y,goldstone)
```

```python
# Contour plot: zdir, offset
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x,y = np.meshgrid(np.arange(-10,10,0.5), np.arange(-10,10,0.5))
r = np.linalg.norm([x,y], axis=0)
goldstone = -200*np.power(r,2)+np.power(r,4)
ax.contour(x,y,goldstone, zdir='y', offset=10)
ax.contour(x,y,goldstone, zdir='x', offset=-10)
ax.contour(x,y,goldstone, zdir='z', offset=-10000)
ax.scatter(x,y,goldstone, c=goldstone, cmap='inferno')
```
