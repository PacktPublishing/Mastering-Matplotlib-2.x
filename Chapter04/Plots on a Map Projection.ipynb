```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

```python
def setup_map(m):
    m.drawcoastlines(zorder=-1)
    m.fillcontinents(color='coral',lake_color='aqua', zorder=-1)
    # draw parallels and meridians.
    m.drawparallels(np.arange(-90.,120.,30.))
    m.drawmeridians(np.arange(0.,420.,60.))
    m.drawmapboundary(fill_color='aqua')
    plt.gcf().set_size_inches(12,10)
```

```python
from mpl_toolkits.basemap import Basemap
```

```python
# Projecting with and without latlon
m = Basemap(width=1.2e7,height=9e6,projection='lcc', lat_1=45.,lat_2=55,lat_0=49.5,lon_0=8.7)
setup_map(m)
HD = (8.7, 49.5)
m.scatter(*HD, latlon=True, color='k', marker='x')
plt.show()
```

```python
# Plotting lines
m = Basemap(width=1.2e7,height=9e6,projection='lcc', lat_1=45.,lat_2=55,lat_0=49.5,lon_0=8.7)
setup_map(m)
m.plot(np.linspace(-10,60,20), 45*np.ones(20), 'g-', latlon=True, lw=3)
m.plot(np.linspace(-10,60,2), 45*np.ones(2), 'k-', latlon=True, lw=3)
plt.show()
```

```python
# The shortest distance: the great circle
m = Basemap(width=1.2e7,height=9e6,projection='cyl', lat_1=45.,lat_2=55,lat_0=49.5,lon_0=8.7)
setup_map(m)
m.drawgreatcircle(-10,45,60,45, color='g', lw=3, linestyle='--')
m.drawgreatcircle(0,50,0,0, color='b', lw=3, linestyle='--')
m.plot(np.linspace(-10,60,20), 45*np.ones(20), 'g-', latlon=True, lw=3)
m.plot(np.linspace(0,0,20), np.linspace(50,0,20), 'b-', latlon=True, lw=3)
plt.show()
```

```python
# Terminator 2: Judgement Datetime (nightshade)
from datetime import datetime
m = Basemap(projection='ortho', lat_0=45, lon_0=45)
setup_map(m)
m.nightshade(datetime.utcnow())
plt.show()
```
