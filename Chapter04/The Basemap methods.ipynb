```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

```python
def setup_map(m):
    m.drawcoastlines()
    m.fillcontinents(color='coral',lake_color='aqua')
    # draw parallels and meridians.
    m.drawparallels(np.arange(-90.,120.,30.))
    m.drawmeridians(np.arange(0.,420.,60.))
    m.drawmapboundary(fill_color='aqua')
    for y in np.linspace(m.ymax/22,19*m.ymax/22,9):
        for x in np.linspace(m.xmax/22,19*m.xmax/22,12):
            lon, lat = m(x,y,inverse=True)
            m.tissot(lon+5,lat+5,1.5,100,facecolor='k',zorder=10,alpha=0.5)
```

```python
from mpl_toolkits.basemap import Basemap
```

```python
# Creating the most basic projection
m = Basemap(llcrnrlat=-80,urcrnrlat=80,llcrnrlon=-180,urcrnrlon=180,lat_ts=0)
setup_map(m)
plt.show()
```

```python
# Mercator projection
m = Basemap(projection='merc',llcrnrlat=-80,urcrnrlat=80,llcrnrlon=-180,urcrnrlon=180,lat_ts=0)
setup_map(m)
plt.show()
```

```python
# Equal Area Projection
m = Basemap(projection='hammer',lon_0=0)
setup_map(m)
plt.show()
```

```python
# Just plotting a region
m = Basemap(width=5e6,height=4e6,projection='lcc', lat_1=45.,lat_2=55,lat_0=49.5,lon_0=8.7, resolution='l')
m.drawcoastlines()
m.fillcontinents(color='coral',lake_color='aqua')
# draw parallels and meridians.
m.drawparallels(np.arange(-90.,120.,30.))
m.drawmeridians(np.arange(0.,420.,60.))
m.drawmapboundary(fill_color='aqua')
plt.show()
```

# Space(!) projections

```python
# Orthographic projection
m = Basemap(projection='ortho',lon_0=10,lat_0=20)
setup_map(m)
plt.show()
```

```python
# Geostationary projection
m = Basemap(projection='geos',lon_0=-79.5)
setup_map(m)
plt.show()
```

```python
# Near-Sided  perspective projection
m = Basemap(projection='nsper',lon_0=-114, lat_0=51, satellite_height=4e5, resolution='l')
setup_map(m)
plt.show()
```
