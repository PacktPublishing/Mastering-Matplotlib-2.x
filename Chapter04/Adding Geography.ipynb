```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (8,6)
plt.rcParams['figure.dpi'] = 150
```

```python
# Just coastlines
m = Basemap(projection='ortho',lon_0=-114, lat_0=51)
m.drawcoastlines()
plt.show()
```

```python
# Resolution
m = Basemap(projection='ortho',lon_0=-114, lat_0=51, resolution='i')
m.drawcoastlines()
plt.show()
```

```python
# Countries & states
m = Basemap(projection='ortho',lon_0=-114, lat_0=51)
m.drawcoastlines()
m.drawcountries()
m.drawstates()
plt.show()
```

```python
m = Basemap(projection='ortho',lon_0=-114, lat_0=51)
m.drawcoastlines()
m.drawrivers(color='b')
m.fillcontinents(color='none', lake_color='b')
plt.show()
```

```python
# Shaded relief
m = Basemap(projection='ortho',lon_0=-114, lat_0=51)
m.shadedrelief(scale=1)
plt.show()
```

```python
# NOAA etopo relief map
m = Basemap(projection='ortho',lon_0=-114, lat_0=51)
m.etopo()
plt.show()
```

```python
# Blue marble
m = Basemap(projection='ortho',lon_0=-114, lat_0=51)
m.bluemarble()
plt.show()
```
