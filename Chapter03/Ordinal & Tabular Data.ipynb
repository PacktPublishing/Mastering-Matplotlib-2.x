```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

# Pie Charts

```python
# Basic Pie Chart
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
plt.pie(fracs, labels=labels);
plt.gca().set_aspect('equal')
```

```python
# Explode!
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
explode = (0, 0.2, 0)
plt.pie(fracs, explode=explode, labels=labels);
plt.gca().set_aspect('equal')
```

```python
# colors
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
plt.pie(fracs, colors=('k', 'c', 'g'), labels=labels);
plt.gca().set_aspect('equal')
```

```python
# Appearance tweaks (shadow, radius, colors)
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
plt.pie(fracs, labels=labels, radius=0.5)
plt.gca().set_aspect('equal')
```

```python
# Show percentages w/ autopct
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
plt.pie(fracs, labels=labels, autopct="%2.0f %%", pctdistance=0.5)
plt.gca().set_aspect('equal')
```

# Tables

```python
# Add a table
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
plt.pie(fracs, labels=labels)
plt.table(cellText=[fracs], colLabels=labels, rowLabels=["fraction"])
plt.gca().set_aspect('equal')
```

```python
# Change location w/ loc
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
plt.pie(fracs, labels=labels)
plt.table(cellText=[fracs], rowLabels=['Fraction'], colLabels=labels, loc='left')
plt.gca().set_aspect('equal')
```

```python
# Cell colours w/ rowColours, cellColours & colColours
fracs = (78, 21, 1)
labels = ('Nitrogen', 'Oxygen', 'Argon')
plt.pie(fracs, labels=labels)
plt.table(cellText=[fracs], rowLabels=['Fraction'], colLabels=labels, cellColours=[['b','g','r']])
plt.gca().set_aspect('equal')
```
