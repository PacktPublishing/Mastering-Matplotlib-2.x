```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

# Polar Axes

```python
nums = np.arange(0,10,0.1)
plt.subplot(111, projection='polar')
plt.plot(nums, nums/10.)
plt.plot(nums, np.sin(nums))
plt.plot(nums, np.cos(nums))
```

# Axis Scales

```python
# Log axes
plt.plot(nums, np.power(10,10*nums))
plt.gca().set_yscale('log')
plt.gca().grid(True)
```

```python
# Symlog axes
plt.plot(nums, np.power(10,nums)-100)
plt.gca().set_yscale('symlog')
plt.gca().grid(True)
```

```python
# Logistic axes
plt.plot(nums, nums/10)
plt.gca().set_yscale('logit')
plt.gca().grid(True)
plt.gca().yaxis.set_minor_formatter(mpl.ticker.NullFormatter())
```
