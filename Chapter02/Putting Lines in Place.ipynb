```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# Set up figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 150
```

```python
# Adding a horizontal and vertical line
nums = np.arange(0,10,0.1)
plt.plot(nums, np.sin(nums))
plt.axhline(0.5, color='r')
plt.axvline(np.pi/2., color='r', linestyle='--')
```

```python
# Adding a horizontal and vertical span
nums = np.arange(0,10,0.1)
plt.plot(nums, np.sin(nums))
plt.axhspan(-0.5,0.5, color='k', alpha=0.5)
```

```python
# Span Axis coverage (min/max)
nums = np.arange(0,10,0.1)
plt.plot(nums, np.sin(nums))
plt.axhspan(-0.5,0.5, xmin=0.5, xmax=0.75, alpha=0.5)
```

```python
# Gridlines
nums = np.arange(0,10,0.1)
plt.plot(nums, np.sin(nums))
plt.grid()
```

```python
# Grid style
nums = np.arange(0,10,0.1)
plt.plot(nums, np.sin(nums))
plt.grid(color='r', linestyle='-')
```
